```python
api_key = "6866d497-d3db-44aa-a5f5-f864ae072926:fx"
url = "https://api-free.deepl.com/v2/translate"
```

```python
def translate_text(text, target_lang="ES"):  # Target language set to Spanish ("ES")
    data = {
        "auth_key": api_key,
        "text": text,
        "target_lang": target_lang
    }
    response = requests.post(url, data=data)
    try:
        translated_text = response.json()['translations'][0]['text']
        return translated_text
    except:
        # Return original text if translation fails
        return text

df['Summary_Spanish'] = df['Summary'].apply(lambda x: translate_text(x))
```
